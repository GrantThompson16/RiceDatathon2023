# Getting a Better Grasp

Previously, we had not realized the scope of the misalligned data. We tried to manually realign it to no avail. We realized that the data in the 'Unnamed 12' column was shifted 4 places over, the data on the 'Unnamed 11' column was shifted 3 places over, and so on. To fix this issue all at once, e decided to move onto different cleaning methods that would be able to shift the data according to the values of the unnamed columns.

```python
#We created a dictionary that maps each column name to its number

shifted_fda = fda.fillna('NA')

shifted_fda.columns

cols_map = {0 :'Facility Name', 1: 'Address 1', 2: 'Address 2', 3: 'Address 3', 4: 'City', 5: 'State',
       6: 'Zip Code', 7: 'Phone', 8: 'Fax', 9: 'Unnamed: 9', 10: 'Unnamed: 10', 11: 'Unnamed: 11',
       12: 'Unnamed: 12'}
```

```python
#We split up our dataframe into two. 
#One of them is misaligned data, which is any row that does not have 'NA' in the 'Unnamed 9' column.
#The other is a set of properly aligned

fda_good = shifted_fda[shifted_fda['Unnamed: 9'] == 'NA']
fda_good

shifted_fda = shifted_fda[shifted_fda['Unnamed: 9'] != 'NA']
shifted_fda
```

```python
#We also realized that the 'City' column is where the misalignment begins. Any data left of City is properly
#aligned, and any data right is misaligned.

shifted_fda_left = shifted_fda[['Facility Name','Address 1','Address 2','Address 3']]
shifted_fda_right = shifted_fda.drop(columns = ['Facility Name','Address 1','Address 2','Address 3'])
```

```python
#We split up our misaligned dataframe further, and used our dictionary to shift the columns the proper amount

shifted_fda_right_12 = shifted_fda_right[shifted_fda_right['Unnamed: 12'] != 'NA'].shift(periods = -4, axis = 1)
shifted_fda_right = shifted_fda_right[shifted_fda_right['Unnamed: 12'] == 'NA']

shifted_fda_right_11 = shifted_fda_right[shifted_fda_right['Unnamed: 11'] != 'NA'].shift(periods = -3, axis = 1)
shifted_fda_right = shifted_fda_right[shifted_fda_right['Unnamed: 11'] == 'NA']

shifted_fda_right_10 = shifted_fda_right[shifted_fda_right['Unnamed: 10'] != 'NA'].shift(periods = -2, axis = 1)
shifted_fda_right = shifted_fda_right[shifted_fda_right['Unnamed: 10'] == 'NA']

shifted_fda_right_9 = shifted_fda_right[shifted_fda_right['Unnamed: 9'] != 'NA'].shift(periods = -1, axis = 1)

```

```python
#We combined all of our right shifted data back into one dataframe
shifted_fda_right = pd.concat([shifted_fda_right_9, shifted_fda_right_10, shifted_fda_right_11, shifted_fda_right_12])
shifted_fda_right
```

```python
#We combined the right and left sides of the dataframes back together
shifted_fda = pd.concat([shifted_fda_left,shifted_fda_right], axis =1)

shifted_fda
```
